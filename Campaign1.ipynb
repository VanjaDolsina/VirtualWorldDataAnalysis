```python
F=[]
for i in range(10):
    Y=[] 
    for node in X[i]: 
        d = X[i].degree(node)
        Y.append([node,d])
    F.append(Y)
```

```python
S=[]
d=[]
r=[]
for i in range(10):
    V= G.subgraph(X[i])
    deg=nx.degree(V)
    pos = nx.spring_layout(V)
    plt.figure(1,figsize=(8,8)) 
    nx.draw_networkx_edges(V,pos, edge_color='k',arrows=True)
    #nx.draw_networkx_nodes(V,pos, node_size=10,node_color='r')
    nx.draw_networkx_nodes(V, pos,nodelist=[p[0] for p in F[i]],node_size=[v * 3 for v in [p[1] for p in F[i]]],node_color='r')
    if(i==0):
        plt.title('2010-10-31 15:04:00 ---- 2010-11-01 15:04:00')
    else:
        plt.title('2010-11-0'+str(i)+' 15:04:00 ---- 2010-11-0'+str(i+1)+' 15:04:00')
    plt.axis('off')
    plt.show()
    S.append(max(nx.connected_component_subgraphs(X1[i]),key=len))
    print("Number of nodes in graph: %d" %V.order())
    d.append(nx.diameter(S[i]))
    print("Diameter:", d[i])
    r.append(nx.radius(S[i]))                         
    print("Radius:", r[i])
```

```python
D=[]
D.append(16453)
for i in L:
    for j in L:
        if(i!=j):
            if(i not in D):
                D.append(i)
            if(j not in D):
                D.append(j)
            if(nx.has_path(X[0],i,j)==True):
                for k in nx.shortest_path(X[0],i,j):
                    if(k not in D):
                        D.append(k)
            else:
                continue            
print(len(D))
print(D)
```
