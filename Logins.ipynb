```python
y=pd.read_csv('logins.csv',sep=";",header=None,engine='python',usecols=(0,1))
```

```python
y
```

```python
s = pd.Series(pd.date_range('2007-11-26 09:40:00','2012-10-05 07:06:24',freq='1M'))
```

```python
s[58]
```

```python
y[0]=y[0].as_matrix() 
```

```python
y[0]= pd.to_datetime(y[0])
```

```python
p=y.groupby(1).count()
```

```python
p.rename(columns={ 0:'User number of logins'})
p.sort_values(0,ascending=False)
```

```python
Time=[]
for i in range(58):
    Time.append(y.loc[(y[0]>=s[i])&(y[0]<s[i+1])])
    print(str(i)+" Total logins from "+s[i].strftime('%Y-%m-%d')+" to "+s[i+1].strftime('%Y-%m-%d')+": "+format(len(Time[i])))
```

```python
plt.figure(figsize=(10,5), dpi=80)
for i in range(58):
    if((i==35)|(i==47)|(i==50)):
        plt.scatter(s[i],len(Time[i]),c='r')
    else:
        plt.scatter(s[i],len(Time[i]),c='c')
plt.grid()
plt.ylabel('Number of logins')
plt.title('Number of logins monthly from 2007 to 2013')
plt.savefig('log1.png')
plt.show()
```
