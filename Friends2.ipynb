```python
y=pd.read_csv('friends.csv',sep=";",header=None,engine='python',usecols=(0,1,2))
```

```python
G=nx.from_pandas_dataframe(y,source=1, target=2,edge_attr=True,create_using=nx.DiGraph())
```

```python
G.order()
```

```python
Y=[]
Y1=[]
Y2=[]
for node in G:
    d = G.degree(node)
    Y.append([node,d])
    d1 = G.out_degree(node)
    Y1.append([node,d1])
    d2 = G.in_degree(node)
    Y2.append([node,d2])
L=[]    
L1=[]
L2=[]
Y=sorted(Y, key=lambda x: x[1], reverse=True)
Y1=sorted(Y1, key=lambda x: x[1], reverse=True)
Y2=sorted(Y2, key=lambda x: x[1], reverse=True)
Y=Y[:1000]
Y1=Y1[:1000]
Y2=Y2[:1000]
#print(Y)
#print(Y1)
#print(Y2)
for u in [p[0] for p in Y]:
    L.append(u)
print("\nUsers with most degrees: ")
print(L[:20])
for u in [p[0] for p in Y1]:
    L1.append(u)
print("\nUsers with most out_degrees: ")
print(L1[:20])
for u in [p[0] for p in Y2]:
    L2.append(u)
print("\nUsers with most in_degrees: ")
print(L2[:20])
```

```python
S = G.subgraph(L)
```

```python
pageranks = nx.pagerank(S)  
sort = sorted(pageranks.items(), key=operator.itemgetter(1),reverse=True)
print('Pageranks:', sort)
```
